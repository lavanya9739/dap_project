```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_data.csv")
```

```python
df.tail()
```

```python
df.head()
```

```python
df
```

```python
print("Basic Statistics:\n")
print(df[['Age', 'GPA']].describe())
```

```python
print("\nNumPy Statistics:")
print("Mean Age:", np.mean(df['Age']))
print("Median Age:", np.median(df['Age']))
print("Standard Deviation of GPA:", np.std(df['GPA']))
```

```python
grade_counts = df['Grade'].value_counts()
print("\nGrade Distribution:\n", grade_counts)
```

```python
major_gpa = df.groupby('Major')['GPA'].mean().sort_values(ascending=False)
print("\nAverage GPA per Major:\n", major_gpa)
```

```python
city_counts = df['City'].value_counts()
print("\nCity-wise Student Count:\n", city_counts)
```

```python
pivot_gpa = df.pivot_table(values='GPA', index='Grade', columns='Major', aggfunc='mean')
print("\nPivot Table - Average GPA by Grade and Major:\n", pivot_gpa)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=grade_counts.index, y=grade_counts.values, palette='viridis')
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=major_gpa.index, y=major_gpa.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Average GPA by Major")
plt.xlabel("Major")
plt.ylabel("Average GPA")

```

```python
plt.subplot(2, 2, 3)
sns.histplot(df['Age'], bins=10, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python

plt.subplot(2, 2, 4)
sns.boxplot(x=df['Grade'], y=df['GPA'], palette='pastel')
plt.title("GPA Distribution by Grade")
plt.xlabel("Grade")
plt.ylabel("GPA")
```

```python

plt.tight_layout()
plt.show()
```
